```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.optimize import curve_fit
from astropy import constants as const
from astropy import units as u
from astropy.io import fits
```

```python
file_names = ["M13_0001.fits", "M13_0002.fits", "M13_0003.fits", "M13_0004.fits", "M13_0005.fits"]

all_data = []

for file in file_names:
    with fits.open(file) as hdul:
        header = hdul[0].header # Acessing metadata information
        data = hdul[0].data
        all_data.append(data)
```

```python
fig, axes = plt.subplots(1,5, figsize = (20,4))

for i, ax in enumerate(axes):
    ax.imshow(all_data[i], origin="lower", cmap="grey")
    ax.set_title(f"Image {i+1}")

plt.show()
```

```python
vmin = min(np.min(img) for img in all_data)
vmax = max(np.max(img) for img in all_data)
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i, ax in enumerate(axes):
    im = ax.imshow(all_data[i], origin="lower", cmap="gray", vmin=vmin, vmax=vmax)
    ax.set_title(f"Image {i+1}")
    

fig.colorbar(im, shrink=0.8)


plt.show()
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i, ax in enumerate(axes):
    ax.hist(all_data[i].ravel(), bins=50, color='gray')
    ax.set_title(f"Image {i+1}")
    ax.set_xlabel("Pixel Intensity")
    ax.set_ylabel("Count")

plt.show()
```

```python
#2 
print (" X axis is pixel intensity (brightness of each pixel), Y axis is number of pixels that have that brightness")
print ("Bins = 50 divides the range of pixel intensities into 50 equal eifth groups")
print ("Most of the data is centered at ~500 pixel intensity")
print ("Ignoring outliers, the data is pretty narrow and not spread out a lot")

for i, img in enumerate(all_data):
    print(f"Image {i+1}: mean = {np.mean(img):.2f}, median = {np.median(img):.2f}")
```

```python
combined_image = np.median(all_data, axis=0)
print("Combined median:", np.median(combined_image))
```
