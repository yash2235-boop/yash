## 3.2

```python
matrix = [
[5, 12, 17, 9],
[8, 21, 14, 3],
[6, 13, 19, 11]
]

print ("shape: ", np.shape(matrix))

print ("third column: ")
for row in matrix:
    print (row[2])

print ("second row: ", matrix[1])

matrix_arr = np.array (matrix)
matrix_2 = matrix_arr * 2
print ("matrix elements times 2: \n ", matrix_2)

print ("mean in each column: ")
col_mean = np.mean(matrix, axis=0)
print(col_mean)

overall_mean = np.mean(matrix)
print("Overall mean of the array:", overall_mean)

max_in_rows = np.max(matrix, axis=1)
min_in_cols = np.min(matrix, axis=0)
print("Max value in each row:", max_in_rows)
print("Min value in each column:", min_in_cols)

flat = matrix_arr.flatten()
print("Flattened array:", flat)
print("New shape:", flat.shape)
print("The 2D array became 1D - all elements are in a single row")

min_index = np.argmin(flat)
max_index = np.argmax(flat)
print("Flattened index of smallest value:", min_index)
print("Flattened index of largest value:", max_index)

min_coords = np.unravel_index(min_index, matrix_arr.shape)
max_coords = np.unravel_index(max_index, matrix_arr.shape)
print("Smallest value coordinates:", min_coords)
print("Largest value coordinates:", max_coords)

greater_than_15 = matrix_arr[matrix_arr > 15]
print("Values greater than 15:", greater_than_15)

sorted_matrix = np.sort(matrix_arr, axis=1)
print("Each row sorted:\n", sorted_matrix)
```

## 4

```python
arr = np.array([
[2, 3, 5],
[14, 6, 8],
[11, 13, 17],
[7, 10, 13]
])

def row_with_prime (arr):
    result = []
    for row in arr:
        for i in row:
            if i > 1 and all (i % n != 0 for n in range (2,int(i**0.5) +1)):
                result.append(row)
                break
    return np.array (result)

print ("Rows with atleast one prime number: ", row_with_prime(arr))
```

## 5.2

```python
def create_checkered_pattern (empty_board):
    for i in range (0 , 8 , 2 ):
        empty_board[i, ::2] = 1
    checkered_board = empty_board
    return checkered_board

almost_complete = create_checkered_pattern(empty_board)

print (almost_complete)


```

## 5.3


```python
def complete_checkered_pattern (almost_complete):
    for i in range (1 ,8, 2):
        almost_complete[i, 1::2] = 1
    completed_board = almost_complete
    return completed_board

print (complete_checkered_pattern(almost_complete))
```

## 5.4

```python
def reversed_checkered ():
    board = np.zeros((8, 8))
    board[::2, 1::2] = 1   
    board[1::2, ::2] = 1   
    return board

print (reversed_checkered())
```

# 6

```python
universe = np.array(['galaxy', 'clusters'])

def expansion (words, n):
    result = []
    for word in words:
        space = " " * n               
        new_word = space.join(word)    
        result.append(new_word)
    return np.array(result)

print (expansion(universe, 5))
```

```python
np.random.seed(123)
stars = np.random.randint(500, 2000, (5, 5))
print (stars)

def second_max(stars):
    
    sorted_rows = np.sort(stars, axis=1)
    return sorted_rows[:, -2]

print ("second max stars: ", second_max(stars))
```
