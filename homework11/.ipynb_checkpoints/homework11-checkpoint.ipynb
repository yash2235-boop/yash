```python
data2 = Table.read("global_CCl4_MM.dat", format = "ascii")
df = data2.to_pandas()
df.columns
```

```python
df = df[['CCl4ottoyr', 'CCl4ottoGLm', 'CCl4ottoGLsd']]
df.head()

```

```python
plt.figure(figsize=(10,5))

plt.errorbar(df['CCl4ottoyr'], df['CCl4ottoGLm'], yerr=df['CCl4ottoGLsd'],
             markersize=5, fmt='o', capsize=2)

plt.xlabel("Year")
plt.ylabel("Global Mean CO_2 Concentration (ppm)")
plt.title("CO_2 Concentration Over Time")
plt.grid(True)
plt.show()
```

```python

df = df.dropna()

def linear_model(x, m, b):
    return m*x + b


x = df['CCl4ottoyr'].values
y = df['CCl4ottoGLm'].values
sigma = df['CCl4ottoGLsd'].values


popt, pcov = curve_fit(linear_model, x, y, sigma=sigma, absolute_sigma=True)

m, b = popt
m_err, b_err = np.sqrt(np.diag(pcov))

y_fit = linear_model(x, m, b)
```

```python
N = len(y)
chi2 = np.sum(((y - y_fit)**2) / sigma**2)
chi2_red = chi2 / (N - 2)
chi2_red
```

```python
plt.figure(figsize=(10,5))

plt.errorbar(x, y, yerr=sigma, fmt='o', markersize =3, capsize=2, label='Data')
plt.plot(x, y_fit, color = "green",  label=f'Fit: y = {m:.2f}x + {b:.1f}')
plt.xlabel("Year")
plt.ylabel("Global Mean Concentration (ppt)")
plt.title("Linear Fit to Global Mean Concentration")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# yes the model is a good fit as the line of best fit passes through all points or their error bars
```
