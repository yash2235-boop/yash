```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("exoplanet_data.csv", comment = "#")
df
```

```python
print ("Shape: ", np.shape(df), "564 rows and 92 columns")
print("Column names: \n", df.columns)
print ("Five columns that I have no idea what they mean: \n sy_kmagerr2 \n pl_insolerr1 \n pl_orbeccenerr2 \n pl_orbpererr2 \n pl_controv_flag")
```

```python
# Planet name - pl_name - Planet name most commonly used in the literature - Gives the name of the exoplanet
 
# Host name - hostname - Stellar name most commonly used in the literature - Name of the star the planet revoles around

# Discovery Method - discoverymethod - Method by which the planet was first identified - Methoid used like transiting, radial velocity, microlensing etc.

# Spectral Type - st_spectype - Classification of the star based on their spectral characteristics following the Morgan-Keenan system - Spectral type of the host star - O, B , A etc.

# Equilibrium Temperature - pl_eqt - The equilibrium temperature of the planet as modeled by a black body heated only by its host star,
# or for directly imaged planets, the effective temperature of the planet required to match the measured luminosity if the planet were a black body - Equilibrium temperatue of the planet in kelvin.

# Planet Mass or Mass*sin(i) [Jupiter Mass] - pl_bmassj - Best planet mass estimate available, in order of preference: Mass, M*sin(i)/sin(i), or M*sin(i), depending on availability, and measured in Jupiter masses 
# - Planet mass or minimum mass in Jupiter masses
```

```python
reduced_columns = ["pl_name", "hostname", "discoverymethod", "st_spectype", "pl_eqt", "pl_bmassj"]

df_reduced = df[reduced_columns]

df_reduced = df_reduced.dropna()

df_reduced

print ("Shape: ", np.shape(df_reduced), "67 rows and 6 columns")
print ("First five rows: \n \n", df_reduced.head(5))
print ("Last five rows: \n \n", df_reduced.tail(5))

print (" \n Column titles: \n", df_reduced.columns)
```

```python
df_reduced.value_counts("discoverymethod")

max_count = np.max(df_reduced.value_counts("discoverymethod"))

```

```python
"""
Discovery Method Analysis
 
The Transit method is the most common with 52 planets, while Imaging is the least common with 2 planets.  
Each method uses a different technique:
- Transit: detects dips in brightness when a planet crosses its star.
- Radial Velocity: detects the star’s wobble due to the planet’s gravity.
- Imaging: directly captures images of exoplanets.
"""
```

```python
plt.figure (figsize = (8,6))

sns.scatterplot (data = df_reduced, x = "pl_bmassj", y = "pl_eqt" , hue = "discoverymethod", s=80)
plt.title("Planet Mass vs Equilibrium Temperature by Discovery Method", fontsize=14)
plt.xlabel("Planet Mass (Jupiter Masses)", fontsize=12)
plt.ylabel("Equilibrium Temperature (K)", fontsize=12)

plt.legend(title="Discovery Method")

plt.show()
```

```python
plt.figure (figsize = (8,6))

sns.scatterplot (data = df_reduced, x = "pl_bmassj", y = "pl_eqt" , hue = "discoverymethod", s=80)

plt.xscale("log")
plt.yscale("log")
plt.title("Planet Mass vs Equilibrium Temperature by Discovery Method", fontsize=14)
plt.xlabel("Planet Mass (Jupiter Masses)", fontsize=12)
plt.ylabel("Equilibrium Temperature (K)", fontsize=12)

plt.legend(title="Discovery Method")

plt.show()
```

```python
# Exoplanets discovered by the imaging method are much more massive. This makes sense because imaging can only detect large, bright planets far from their stars, which are easier to capture directly.
```

```python
massive_planets = df_reduced[df_reduced["pl_bmassj"] > 5]
coldest_planet = massive_planets.nsmallest(1, "pl_eqt")
print(coldest_planet)


index = coldest_planet.index[0]
name = coldest_planet["pl_name"].values[0]
temp = coldest_planet["pl_eqt"].values[0]
mass = coldest_planet["pl_bmassj"].values[0]
spectral_type = coldest_planet["st_spectype"].values[0]

print(f"\nIndex: {index}")
print(f"Planet Name: {name}")
print(f"Equilibrium Temperature: {temp} K")
print(f"Mass: {mass} Jupiter Masses")
print(f"Host Star Spectral Type: {spectral_type}")

if isinstance(spectral_type, str) and spectral_type.startswith("G"):
    print("This planet orbits a Sun-like star (G-type).")
else:
    print("This planet does NOT orbit a Sun-like star.")
```

```python
plt.figure (figsize = (8,6))

sns.scatterplot (data = df_reduced, x = "pl_bmassj", y = "pl_eqt" , hue = "discoverymethod", s=80)

plt.scatter(coldest_planet["pl_bmassj"], coldest_planet["pl_eqt"], color="red", s=250, marker="*", label=coldest_planet["pl_name"].values[0])

plt.xscale("log")
plt.yscale("log")
plt.xlabel("Planet Mass (Jupiter Masses)", fontsize=12)
plt.ylabel("Equilibrium Temperature (K)", fontsize=12)
plt.title("Planet Mass vs Equilibrium Temperature (Highlighted Coldest Planet)", fontsize=14)
plt.legend(title="Discovery Method + Coldest Planet")

plt.show()
```
