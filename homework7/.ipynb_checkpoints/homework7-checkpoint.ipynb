```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2.1

```python
def make_sine_wave(x, A, w):
    y = A * np.sin (w *x)
    return y
```

### 2.2

```python
pi = np.pi
x = np.linspace (0,2*pi, 1000)
```

### 2.3

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

### 2.4

```python
fig, axes = plt.subplots(1, 3, figsize=(25, 5))
```

### 2.5

```python
for A, w in zip(amplitude, frequency):
    print(f"Amplitude = {A}, Frequency = {w}")
```

### 2.6

```python
fig, axes = plt.subplots(1, 3, figsize=(25, 5))
#keeping omega constant:
for A in amplitude:
    y = make_sine_wave(x, A, 1)
    axes[0].plot(x, y, label=f"A = {A}", linestyle='--')
axes[0].set_title("Effect of Amplitude on Sine Wave ( = 1)", fontsize=16)
axes[0].set_xlabel("x (radians)", fontsize=14)
axes[0].set_ylabel("y = Asin(x)", fontsize=14)
axes[0].legend(fontsize=12)
axes[0].grid(True)

#keeping A constant:
for w in frequency:
    y = make_sine_wave(x, 1, w)
    axes[1].plot(x, y, label=f" = {w}", linestyle='-.')
axes[1].set_title("Effect of Frequency on Sine Wave (A = 1)", fontsize=16)
axes[1].set_xlabel("x (radians)", fontsize=14)
axes[1].set_ylabel("y = Asin(x)", fontsize=14)
axes[1].legend(fontsize=12)
axes[1].grid(True)

#vary both together
for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w)
    axes[2].plot(x, y, label=f"A = {A},  = {w}")
axes[2].set_title("Changing Both Amplitude and Frequency", fontsize=16)
axes[2].set_xlabel("x (radians)", fontsize=14)
axes[2].set_ylabel("y = Asin(x)", fontsize=14)
axes[2].legend(fontsize=12)
axes[2].grid(True)

plt.show()
```
