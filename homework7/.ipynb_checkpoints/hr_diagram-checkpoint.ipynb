```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)

print (np.shape(data))
#2542 rows, 2 columns
```

```python
magnitude =[]
temperature =[]
for row in data:
    magnitude.append (row[0])

for row in data:
    temperature.append (row[1])




print ("The minimum of the magnitude column is: ", np.min(magnitude), "and corresponds to the brightest star.")
print ("The maximum of the magnitude column is: ", np.max(magnitude), "and corresponds to the dimmest star.")
print ("The minimum of the temperature column is: ", np.min(temperature), "and corresponds to the coolest star.")
print ("The maximum of the temperature column is: ", np.max(temperature), "and corresponds to the hottest star.")
```

## plotting the HR diagram

```python
#indices of key stars
brightest_index = np.argmin(magnitude)
dimmest_index = np.argmax(magnitude)
hottest_index = np.argmax (temperature)
coolest_index = np.argmin (temperature)

#data converted to log
log_magnitude = np.log (temperature)
log_temperature = np.log (magnitude)

#creating 1 plot
fig, axes = plt.subplots(1, 1, figsize=(8, 6))

#plot all points
plt.scatter (log_magnitude, log_temperature,s = 5, color = 'blue')

#plot key stars
plt.scatter (log_magnitude[brightest_index], log_temperature[brightest_index], s = 100, color = "black", label = "Brightest Star", marker = "*")
plt.scatter (log_magnitude[dimmest_index], log_temperature[dimmest_index], s = 100, color = "purple", label = "Dimmest Star", marker = "*")
plt.scatter (log_magnitude[hottest_index], log_temperature[hottest_index], s = 100, color = "green", label = "Hottest Star", marker = "*")
plt.scatter (log_magnitude[coolest_index], log_temperature[coolest_index], s = 100, color = "orange", label = "Coldest Star", marker = "*")

#axes labels
plt.xlabel("Temperature (log K)")
plt.ylabel(" Magnitude ")
plt.title("HertzsprungRussell Diagram")
plt.gca().invert_xaxis()   
plt.gca().invert_yaxis() 

#show stuff
plt.legend ()
plt.show()
```
