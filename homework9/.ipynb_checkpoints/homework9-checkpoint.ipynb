```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import seaborn as sns
import pandas as pd
```

```python
df1 = pd.read_csv("noisy_curvefit_data.csv")
print ("\n Shape: ", df1.shape)
print ("\n1 00 rows, 6 columns")
print("\n First 5 rows: \n \n", df1.head(5))
print("\n Last 5 rows: \n \n", df1.tail(5))
print("\n Column names: \n", df1.columns)
```

```python
x1 = df1["x"]
y1 = df1["y_noise_1"]

plt.figure(figsize = (8,6))
plt.scatter (x1, y1, color = "grey", s = 30, label = "Noisy Data")
plt.xlabel("X data")
plt.ylabel ("Y noisy data")
plt.title("First noise level")

plt.legend()
plt.show ()
```

# 2.2 - Fit the Data

```python
def model (x, A, k):
    return A * np.exp(-k * x)
```

```python
p0 = [1,1]
popt, pcov = curve_fit(model, x1, y1, p0=p0)

A_fit, k_fit = popt

print (f"Fit: y = {A_fit :.2f} * e ^ (-{k_fit :.2f} * x)")
```

```python
residuals = y1 - model (x1, A_fit, k_fit)
chi2 = np.sum (residuals ** 2)
dof = len (x1) - len (popt)
chi2_reduced = chi2 / dof
```

```python

plt.figure(figsize=(8,6))

plt.scatter(x1, y1, label="Noisy data", color="blue", s = 30)

plt.plot(x1, model(x1, A_fit, k_fit), color="red", label="Fitted model")

plt.plot(x1, model(x1, 5, 0.8), 'k--', label="True model")

plt.title(f"Noise level: 1 - Reduced chi")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.show()
```

# 2.3 - Compare Different Noise Levels

```python
def fit_model (x,y):
    p0 = [1,1]
    popt, pcov = curve_fit(model, x, y, p0=p0)
    A_fit, f_fit = popt
    residuals = y - model (x, A_fit, k_fit)
    chi2 = np.sum (residuals ** 2)
    dof = len (x) - len (popt)
    chi2_reduced = chi2 / dof

    return A_fit, k_fit, chi2_reduced
```

```python
fig, axes = plt.subplots (2, 2, figsize = (10,8))
axes = axes.flatten() 

noise_cols = ['y_noise_2', 'y_noise_3', 'y_noise_4', 'y_noise_5']

for i, col in enumerate (noise_cols):
    y = df1[col]
    A_fit, k_fit, chi2_red = fit_model(x1, y)

    ax = axes[i]
    ax.scatter (x1, y, color = "gray", label = "Noisy data")
    ax.plot(x1, model(x1, A_fit, k_fit), color='red', label='Fitted model')
    ax.plot(x1, model(x1, 5, 0.8), 'k--', label='True model')

    ax.set_title(f"{col}: Reduced chi = {chi2_red:.2f}")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.legend()

plt.show()

#4
# As noise increases, the red fitted line deviates  more from the black line as teh data becomes too scattered
# The reduced chi squared increases as with noise as larger residuals make the fit less accurate
```

# 2.4 - Save your results

```python
noise_cols = ['y_noise_1', 'y_noise_2', 'y_noise_3', 'y_noise_4', 'y_noise_5']

results = []

for col in noise_cols:
    y = df1[col]
    A_fit, k_fit, chi2_red = fit_model(x1, y)
    results.append([col, A_fit, k_fit, chi2_red])

df_results = pd.DataFrame(results, columns=['Noise level', 'A fit', 'k fit', 'Reduced chi-squared'])

df_results.to_csv("noisy-parameters.csv", index=False)

print(df_results)
```
